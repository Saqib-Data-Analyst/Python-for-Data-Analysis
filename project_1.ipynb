# Data Collection & Cleaning

```python
import pandas as pd

data = pd.read_csv("Coffe_sales (4).csv")
```

# count and display the total number of duplicate rows

```python
print(data.duplicated().sum())
data = data.drop_duplicates()
```

# handle missing values (remove rows with any NaN)

```python
data = data.dropna()
```

# Correct formatting issues

```python
# Convert date to proper datetime
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Fix time formatting (replace '.' with ':')
data['Time'] = data['Time'].astype(str).str.replace('.', ':', regex=False)

# Standardize capitalization for categorical columns
for col in ['cash_type', 'coffee_name', 'Time_of_Day', 'Weekday', 'Month_name']:
    data[col] = data[col].str.strip().str.title()
```

# View cleaned data

```python
print(data.info())
print(data.head())
```

# Export Clean Dataset 

```python
data.to_csv("Coffee_clean_data.csv")
```
